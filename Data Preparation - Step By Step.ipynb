```python
import numpy as np
import pandas as pd
```

### Read the CSV

```python
df_lappy = pd.read_csv('lappy_flipkart.csv')
df_lappy.head()
```

### Separate the Brand Name from name

```python
df_lappy['Brand'] = np.nan
for i in range(df_lappy.shape[0]):
    df_lappy['Brand'][i] = df_lappy['name'][i].split()[0]
    
df_lappy.head()
```

### Look for missing Values

```python
df_lappy.isnull().sum()/df_lappy.shape[0]
```

### Separate Ratings and No. of Reviews

```python
df_lappy.num_ratings[0]
```

```python
df_lappy.num_ratings[0].replace('\xa0','').split('&')
```

```python
df_lappy['num_reviews'] = np.nan
for i in range(df_lappy.shape[0]):
    if(not pd.isnull(df_lappy['num_ratings'][i])):
        rr = df_lappy['num_ratings'][i].replace('\xa0','').split('&')
        rat_temp = rr[0].split()[0]
        rev_temp = rr[1].split()[0]

        if (',' in rat_temp):
            df_lappy['num_ratings'][i] = int(rr[0].split()[0].replace(',',''))
        else:
            df_lappy['num_ratings'][i] = int(rr[0].split()[0])

        if (',' in rev_temp):
            df_lappy['num_reviews'][i] = int(rr[1].split()[0].replace(',',''))
        else:
            df_lappy['num_reviews'][i] = int(rr[1].split()[0])
    
        

df_lappy.head()
```

### Clean RAM Variable and extract ram size and ram type

```python
list(df_lappy['ram'])
```

```python
df_lappy['ram'][1].split()
```

```python
df_lappy['ram_type'] = np.nan
for i in range(df_lappy.shape[0]):
    ram_split = df_lappy['ram'][i].split()
    df_lappy['ram'][i] = ram_split[0]
    df_lappy['ram_type'][i] = ram_split[2]
```

```python
df_lappy.head()
```

```python
df_lappy.storage.value_counts()
```

### Clean Price Column

```python
for i in range(df_lappy.shape[0]):
    df_lappy['price'][i] = df_lappy['price'][i].replace('₹','').replace(',','')
```

### Clean Original Price Column

```python
pd.isnull(df_lappy['org_price'][0])
```

```python
for i in range(df_lappy.shape[0]):
    if(not pd.isnull(df_lappy['org_price'][i])):
        if(',' in df_lappy['org_price'][i]):
            df_lappy['org_price'][i] = df_lappy['org_price'][i].replace('₹','').replace(',','')
```

```python
df_lappy.head()
```

### Clean The column Display

```python
list(df_lappy['display'])
```

```python
df_lappy.iloc[30:50]
```

```python
for i in range(df_lappy.shape[0]):
    df_lappy['display'][i] = df_lappy['display'][i].split()[0]
```

```python
df_lappy['display'] = pd.to_numeric(df_lappy['display'], errors='coerce')
```

### Convert Price, org_price, ram and num_ratings to numeric

```python
df_lappy['price'] = pd.to_numeric(df_lappy['price'], errors='coerce')
df_lappy['org_price'] = pd.to_numeric(df_lappy['org_price'], errors='coerce')

df_lappy['ram'] = pd.to_numeric(df_lappy['ram'], errors='coerce')
df_lappy['num_ratings'] = pd.to_numeric(df_lappy['num_ratings'], errors='coerce')
```

# Feature Engineering

### Create a new column Discount

```python
df_lappy['Discount'] = np.nan
for i in range(df_lappy.shape[0]):
    df_lappy['Discount'][i] = ((df_lappy['org_price'][i] - df_lappy['price'][i])*100)/df_lappy['org_price'][i]
```
